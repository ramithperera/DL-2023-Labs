```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Au_nanoParticle_dataset.csv")
data.head(20)
```

```python
new_df = data[["N_total", "N_bulk", "N_surface", "R_avg"]]
```

```python
new_df.head(20)
```

```python
feature_stats = new_df.describe()
feature_stats
```

```python
# Plot histograms
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
for i, feature in enumerate(new_df.columns):
    sns.histplot(new_df[feature], kde=True, ax=axes[i])
    axes[i].set_title(f"Histogram Of - {feature}")
plt.show()
```

```python
# Plot pairplots
sns.pairplot(new_df)
plt.show()
```

```python
# New pairplot with customized diagonals and lower half
g = sns.PairGrid(new_df)
g.map_upper(sns.histplot)   # Bivariate histogram
g.map_diag(sns.histplot, kde=True)   # Histogram with KDE
g.map_lower(sns.kdeplot)    # Bivariate kernel density estimation plot
plt.show()
```
